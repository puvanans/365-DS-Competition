### Category extraction

 - The process of discovering engagement categories and their frequency of reptition is repeated for the top 5 represented countries in the coming cells.
 - Goal is study the distribution of these categories and study their similarity.


```python
India = student_engagement_by_country[student_engagement_by_country['country'] == 'IN']

IND_df = pd.DataFrame(India['student_id'].value_counts()).rename(columns = {'student_id':'No,of engagement (IND)'})

IND_df
```

 #### The above cell extracts engagement categories for India

```python
IND_engagement_frequency = pd.DataFrame(IND_df['No,of engagement (IND)'].value_counts()).reset_index().rename(columns = {'index':'Engagement categories','No,of engagement (IND)':'No, of students (IND)'})
IND_engagement_frequency
# .plot()
# (IND_engagement_frequency)
```

#### The above cell extracts the distribution of engagement categories in India

```python
US = student_engagement_by_country[student_engagement_by_country['country'] == 'US']
# Extracts Engagemen data for Us students

US_df = pd.DataFrame(US['student_id'].value_counts()).rename(columns = {'student_id':'No,of engagement(US)'})
# Extracts Student ID (as index) and how often they engage (no, of engagment)  we will consider this to be categories.
US_df
```

```python
US_engagement_frequency = US_df['No,of engagement(US)'].value_counts().reset_index().rename(columns = {'index':'Engagement Categories','No,of engagement(US)':'No,of students(US)'} )
US_engagement_frequency
```

```python
EG = student_engagement_by_country[student_engagement_by_country['country'] == 'EG']
EG_df = pd.DataFrame(EG['student_id'].value_counts()).reset_index().rename(columns = {'student_id':'No,of engagement','index':'Student ID'})
EG_df
```

```python
EG_engagement_frequency = pd.DataFrame(EG_df['No,of engagement'].value_counts()).reset_index().rename(columns ={'index':'Engagement Categories','No,of engagement':'No, of students(EG)'})
EG_engagement_frequency
```

```python
GB = student_engagement_by_country[student_engagement_by_country['country'] == 'GB']
GB_df = pd.DataFrame(GB['student_id'].value_counts()).reset_index().rename(columns = {'student_id':'No,of engagement','index':'Student ID'})

GB_df
```

```python
GB_engagement_frequency = pd.DataFrame(GB_df['No,of engagement'].value_counts()).reset_index().rename(columns = {'index':'Engagement Categories','No,of engagement':'No,of students(GB)'})
GB_engagement_frequency
```

```python
NG = student_engagement_by_country[student_engagement_by_country['country'] == 'NG']
NG_df = pd.DataFrame(NG['student_id'].value_counts()).reset_index().rename(columns = {'student_id':'No,of engagement','index':'Student ID'})
NG_df
```

```python
NG_engagement_frequency = pd.DataFrame(NG_df['No,of engagement'].value_counts()).reset_index().rename(columns={'index':'Engagement Categories','No,of engagement':'No,of students(NG)'})
NG_engagement_frequency
```
