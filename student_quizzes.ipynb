```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import import_ipynb
from Student_info import country_Student
from Student_info import student_info
```

```python
student_quizzes = pd.read_csv('dataSets/365_database/365_student_quizzes.csv')
student_quizzes.head()
```

```python
missing_country_df = student_info[student_info.isnull().any(axis=1)]
missing_country = missing_country_df['student_id'].tolist()
missing_country
```

```python
df = student_quizzes.copy()

df_cleaned = df[~df['student_id'].isin(missing_country)]
student_quizzes_cleaned = df_cleaned.reset_index(drop=True)
(student_quizzes_cleaned)
```

```python
country = []
i = 0
n = len(student_quizzes_cleaned)-1

while (i<=n):
    x=student_quizzes_cleaned['student_id'].loc[i]
    y=country_Student[x]
    country.append(y)
    i=i+1
```

```python
print(len(country))
df_country = pd.DataFrame(country,columns = ['country'])
df_country
```

```python
frames = [student_quizzes_cleaned, df_country]

student_quiz_by_country = pd.concat(frames,axis = 1)

len(student_quiz_by_country['country'].unique())
```
